```python
def obtener_info(s_nota):

    #Creamos un diccionario vacio para probarlo con la informaci√≥n
    ret_dict = {}

    #Extraemos el t√≠tulo del job
    titulo = s_nota.find('h1', attrs={'class':'title'}).find('span')
    if titulo:
      ret_dict['titulo'] = titulo.text
    else:
      ret_dict['titulo'] = ""

    #Extraemos NOC
    NOC = s_nota.find('span', attrs={'noc-no'})
    if NOC:
      ret_dict['NOC'] = NOC.text
    else:
      ret_dict['NOC'] = ""

    #Extraemos la fecha_post
    fecha_post = s_nota.find('span', attrs={'date'})
    if fecha_post:
      ret_dict['fecha_post'] = fecha_post.text
    else:
      ret_dict['fecha_post'] = ""
      
    #Extraemos la compa√±ia
    compania = s_nota.find('span', attrs={'property':'hiringOrganization'})
    if compania:
      ret_dict['compania'] = compania.text
    else:
      ret_dict['compania'] = ""

    #Extraemos locaci√≥n
    locacion = s_nota.find('span', attrs={'property':'addressRegion'})
    if fecha_post:
      ret_dict['locacion'] = locacion.text
    else:
      ret_dict['locacion'] = ""
    
    #Extraemos salario
    salario = s_nota.find('span', attrs={'attribute-value'}).find('span')
    if salario:
      ret_dict['salario'] = salario.text
    else:
      ret_dict['salario'] = ""

    #Extraemos Tipo_contrato
    tipo_contrato = s_nota.find('span', attrs={'property':'employmentType'}).find('span')
    if tipo_contrato:
      ret_dict['tipo_contrato'] = tipo_contrato.text
    else:
      ret_dict['tipo_contrato'] = ""

    #Extraemos vacantes
    vacantes = s_nota.find('span', attrs={'tp_vacancyNumber'})
    if vacantes:
      ret_dict['vacantes'] = vacantes.text
    else:
      ret_dict['vacantes'] = ""
   
    #Extraemos JobBank id
    JobBank_id = s_nota.find('ul',attrs={'job-posting-brief colcount-lg-2'}).find('li')
    if JobBank_id:
      ret_dict['JobBank_id'] = JobBank_id.text
    else:
      ret_dict['JobBank_id'] = ""
   
    #Extraemos Laguages
    idioma = s_nota.find('p', attrs={'property':'qualification'})
    if idioma:
      ret_dict['idioma'] = idioma.text
    else:
      ret_dict['idioma'] = ""

    #Extraemos Educacion
    educacion = s_nota.find('ul', attrs={'property':'educationRequirements qualification'})
    if educacion:
      ret_dict['educacion'] = educacion.text
    else:
      ret_dict['educacion'] = ""

    #Extraemos Experiencia
    experiencia = s_nota.find('p', attrs={'property':'experienceRequirements qualification'})
    if experiencia:
      ret_dict['experiencia'] = experiencia.text
    else:
      ret_dict['experiencia'] = ""

    #Extraemos Responsabilidades
    responsabilidades = s_nota.find('div',attrs={'property':'responsibilities'})
    if responsabilidades:
      ret_dict['responsabilidades'] = responsabilidades.text
    else:
      ret_dict['responsabilidades'] = ""

    return ret_dict
```

```python
def scrape_nota(links_secciones):
  try:
    nota = requests.get(links_secciones)
  except Exception as e:
    print('Error scrapeando URL, error en la request',links_secciones)
    print(e)
    print('\n')
    #return ""

  if nota.status_code != 200:
    print("")#(f'Error obtenido nota {links_secciones}')
    print(f'status code = {nota.status_code}')
    #return ""
  
  s_nota = BeautifulSoup(nota.text, 'lxml')

  ret_dict = obtener_info(s_nota)
  ret_dict['links_secciones'] = links_secciones

  return ret_dict
```
