**Sort and split the dataframe:** You can sort rows using the sort_values method, passing a column name by which you want to sort. Setting the ascending argument to False will sort the data upside down.
*texto en cursiva*
**Column subsets:** You can select a column from the df, using the Dataframe name, followed by square brackets([]) with a column name inside. To select multiple columns, you need two pairs of square brackets (df[["col_a", "col_b"]]). In this code, the inner and outer brackets perform different tasks. The outer brackets are responsible for dividing the Dataframe into subsets, and the inner brackets create a list of column numbers to divide into subsets.
*texto en cursiva*
**Subset of rows:** Sometimes referred to as row filtering or row selection. The most common way to do this is to create a logical condition for filtering. A True or False value is obtained for each row. We can use the logical condition inside the square brackets to create a subset of the rows we are interested in.

**Text-based subset:** we can create subsets of fials based on text data. Here, we use the double equal sign in the logical condition to filter.

**Date-based subsets:** we can create subsets based on dates, using the international standard date format.
